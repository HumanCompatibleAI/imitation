```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

from imitation import bc
from imitation import util
```

```python
env_id = 'CartPole-v1'
policy_dir = '../data/'
env = util.make_vec_env(env_id, 8)
expert_algos = util.load_policy(env, basedir=policy_dir)
if not expert_algos:
    raise ValueError(env)
bc_trainer = bc.BCTrainer(
    env, expert_trainers=expert_algos,
    n_expert_timesteps=4000)
```

```python
bc_trainer.train()
```
